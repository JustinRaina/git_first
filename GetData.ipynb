```python
startprice = tsmc.groupby([tsmc.index.year,tsmc.index.month]).first()
```

```python
endprice = tsmc.groupby([tsmc.index.year,tsmc.index.month]).last()
```

```python
profit = endprice / startprice
```

```python
profit.index.set_names(["year","month"],inplace=True)
```
